```python
## 相関を見る

import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

def analyze_target_variable(df, target_variable, exclude_columns=None, alpha=0.1):
    """
    指定されたターゲット変数に対して、相関分析とLASSO回帰による特徴量選択を行う。

    Parameters:
    - df: pandas.DataFrame
        分析対象のデータフレーム。
    - target_variable: str
        目的変数の列名。
    - exclude_columns: list of str, optional
        分析から除外する列名のリスト。デフォルトは ['都道府県', '地域コード']。
    - alpha: float, optional
        LASSO回帰の正則化パラメータ。デフォルトは 0.1。
    """

    if exclude_columns is None:
        exclude_columns = ['都道府県', '地域コード']

    # データのコピーと前処理
    df_processed = df.copy()

    # 数値変換
    for col in df_processed.columns:
        if col in exclude_columns:
            continue
        if df_processed[col].dtype == 'object':
            df_processed[col] = df_processed[col].str.replace(',', '', regex=False)
            df_processed[col] = pd.to_numeric(df_processed[col], errors='coerce')

    # 数値列の抽出
    numerical_df = df_processed.select_dtypes(include=['number'])

    if target_variable not in numerical_df.columns:
        print(f"Warning: '{target_variable}' 列が見つかりません。列名を確認してください。")
        return

    feature_columns = numerical_df.drop(columns=[target_variable]).columns

    print(f"\n'{target_variable}' と他の数値列との相関係数と無相関検定の結果:")

    # 相関係数とp値の計算
    correlation_results = []
    for col in feature_columns:
        temp_df = df_processed[[target_variable, col]].dropna()
        if len(temp_df) > 1:
            correlation, p_value = pearsonr(temp_df[target_variable], temp_df[col])
            correlation_results.append({'特徴量': col, '相関係数': correlation, 'p値': p_value})
        else:
            correlation_results.append({'特徴量': col, '相関係数': None, 'p値': None})

    correlation_df = pd.DataFrame(correlation_results)
    display(correlation_df.sort_values(by='相関係数', ascending=False))

    # 有意な相関のある特徴量
    alpha_threshold = 0.05
    significant_correlations = correlation_df[correlation_df['p値'] < alpha_threshold].sort_values(by='相関係数', ascending=False)
    print(f"\n有意水準 {alpha_threshold} で有意な相関がある列:")
    display(significant_correlations)

    return significant_correlations
```

## geopandasを使った可視化

```python
# 関数の定義

import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable


def plot_prefecture_heatmap(df, japan_pref, numerator_jp, denominator_jp=None, cmap='Reds'):
    """
    都道府県単位の値を日本地図上にヒートマップで表示する関数

    Parameters:
    - df: データフレーム（都道府県列名は「都道府県」）
    - japan_pref: GeoDataFrame（都道府県地図データ、「N03_001」で都道府県名と結合）
    - numerator_jp: 値として可視化したい日本語列名（例："出生数"）
    - denominator_jp: 割る対象の日本語列名（例："総人口"、指定しなければ割らない）
    - cmap: 色のカラーマップ（デフォルトは赤ベースの "Reds"）
    """

    # データ準備
    df2 = df[['都道府県', numerator_jp]].rename(columns={'都道府県': 'N03_001', numerator_jp: 'value'}).copy()
    df2['value'] = pd.to_numeric(df2['value'], errors='coerce')

    # 分母がある場合は除算
    if denominator_jp:
        denom = df[['都道府県', denominator_jp]].rename(columns={'都道府県': 'N03_001', denominator_jp: 'denom'}).copy()
        denom['denom'] = pd.to_numeric(denom['denom'], errors='coerce')
        df2 = df2.merge(denom, on='N03_001', how='left')
        df2['value'] = df2['value'] / df2['denom']

    # マージ
    merged = japan_pref.merge(df2[['N03_001', 'value']], on='N03_001', how='left')

    # 描画準備
    fig, ax = plt.subplots(figsize=(10, 8))
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.1)

    # 描画
    merged.plot(
        column='value',
        cmap=cmap,
        linewidth=0.5,
        edgecolor='black',
        legend=True,
        categorical=False,
        vmin=merged['value'].min(),
        vmax=merged['value'].max(),
        cax=cax,
        ax=ax
    )

    # 凡例ラベル・タイトル設定
    num_label = label_map.get(numerator_jp, numerator_jp)
    if denominator_jp:
        denom_label = label_map.get(denominator_jp, denominator_jp)
        cax.set_ylabel(f"{num_label} per {denom_label}", fontsize=12)
        ax.set_title(f"{num_label} per {denom_label} by Prefecture", fontsize=16)
    else:
        cax.set_ylabel(num_label, fontsize=12)
        ax.set_title(f"{num_label} by Prefecture", fontsize=16)

    ax.axis('off')
    plt.tight_layout()
    plt.show()
```
